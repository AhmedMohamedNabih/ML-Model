## Model Training (with GridSearchCV)

## Logistic Regression (Tuning) 2 


```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

param_grid = [
    {
        'penalty': ['l1', 'l2'],
        'C': [0.1, 1, 10],
        'solver': ['liblinear']
    },
    {
        'penalty': ['elasticnet'],
        'C': [0.1, 1, 10],
        'solver': ['saga'],
        'l1_ratio': [0.1, 0.5, 0.9]
    }
]

grid = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid,
    cv=5,
    scoring='f1'
)

grid.fit(x_train_scailed, y_train)
```

```python
print("Best Params:", grid.best_params_)
print("Best CV F1 :", grid.best_score_)
```

```python
best_lr = grid.best_estimator_
y_pred = best_lr.predict(x_test_scailed)
print('Accuracy :',accuracy_score(y_test,y_pred))
print('Precision :',precision_score(y_test,y_pred))
print('Recall    :',recall_score(y_test,y_pred))
print('f1 :',f1_score(y_test,y_pred))
```

## Decision Tree (Tuning) 1 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# 1) Grid Ù„Ù„Ù€ hyperparameters Ø§Ù„Ù…Ø·Ù„ÙˆØ¨Ø©
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# 2) GridSearchCV Ø¹Ù„Ù‰ Decision Tree
grid_dt = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring='f1',
    n_jobs=-1
)

grid_dt.fit(x_train, y_train)

# 3) Ø£Ø­Ø³Ù† Ø¨Ø§Ø±Ø§Ù…ØªØ±Ø§Øª + Ø£Ø­Ø³Ù† Score (CV)
print("Best params :", grid_dt.best_params_)
print("Best CV F1  :", grid_dt.best_score_)

# 4) ØªÙ‚ÙŠÙŠÙ… Ø£Ø­Ø³Ù† Ù…ÙˆØ¯ÙŠÙ„ Ø¹Ù„Ù‰ test
best_dt = grid_dt.best_estimator_
y_pred_dt = best_dt.predict(x_test)

print("\n=== Decision Tree (Best Model on Test) ===")
print("Accuracy :", accuracy_score(y_test, y_pred_dt))
print("Precision:", precision_score(y_test, y_pred_dt))
print("Recall   :", recall_score(y_test, y_pred_dt))
print("F1 Score :", f1_score(y_test, y_pred_dt))

# 5) feature_importances_
feat_imp = pd.Series(best_dt.feature_importances_, index=x_train.columns).sort_values(ascending=False)
print("\nTop features:\n", feat_imp.head(10))
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

results = []

# Logistic Regression L2
lr_l2 = LogisticRegression(C=0.1, max_iter=1000)
lr_l2.fit(x_train, y_train)
results.append(["LogReg (L2)",
                accuracy_score(y_train, lr_l2.predict(x_train)),
                accuracy_score(y_test, lr_l2.predict(x_test))])

# Logistic Regression L1
lr_l1 = LogisticRegression(C=0.1, penalty='l1', solver='liblinear', max_iter=1000)
lr_l1.fit(x_train, y_train)
results.append(["LogReg (L1)",
                accuracy_score(y_train, lr_l1.predict(x_train)),
                accuracy_score(y_test, lr_l1.predict(x_test))])

# Decision Tree (Baseline)
dt = DecisionTreeClassifier(random_state=42)
dt.fit(x_train, y_train)
results.append(["Decision Tree",
                accuracy_score(y_train, dt.predict(x_train)),
                accuracy_score(y_test, dt.predict(x_test))])

# Decision Tree (Tuned - Ø¨Ø³ÙŠØ·)
dt_tuned = DecisionTreeClassifier(max_depth=3, min_samples_leaf=5, random_state=42)
dt_tuned.fit(x_train, y_train)
results.append(["Decision Tree (Tuned)",
                accuracy_score(y_train, dt_tuned.predict(x_train)),
                accuracy_score(y_test, dt_tuned.predict(x_test))])

# Random Forest (Baseline)
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)
results.append(["Random Forest",
                accuracy_score(y_train, rf.predict(x_train)),
                accuracy_score(y_test, rf.predict(x_test))])

# Random Forest (Tuned - Ø¨Ø³ÙŠØ·)
rf_tuned = RandomForestClassifier(n_estimators=50, max_depth=3, random_state=42)
rf_tuned.fit(x_train, y_train)
results.append(["Random Forest (Tuned)",
                accuracy_score(y_train, rf_tuned.predict(x_train)),
                accuracy_score(y_test, rf_tuned.predict(x_test))])

pd.DataFrame(results, columns=["Model", "Train Accuracy", "Test Accuracy"])
```
